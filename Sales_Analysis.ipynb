```python
cities = sales_data["City"].unique()
print(cities)
```

```python
product_sells_per_city = {
    city : sales_data[sales_data["City"] == city][["Product", "Quantity_Ordered"]].groupby("Product").sum()["Quantity_Ordered"].sort_values(ascending=False) for city in cities
}
```
